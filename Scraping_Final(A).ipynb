```python
 # Install required libraries
!pip install beautifulsoup4 requests pandas

# Import necessary libraries
import requests  # Library for sending HTTP requests
from bs4 import BeautifulSoup  # Library for parsing HTML documents
import pandas as pd  # Library for data manipulation and analysis
```

```python
# URL for Craigslist NYC Apartments for Rent
base_url = "https://newyork.craigslist.org/d/apartments-housing-for-rent/search/apa"
```

```python
# Send a GET request to the website and store the response
response = requests.get(base_url)
response  # This will show the response object to check if the request was successful
```

```python
# Parse the HTML content of the page with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
soup  # This shows the parsed HTML content
```

```python
# Find all 'div' elements on the page
div = soup.find_all("div")
div  # Displays all the 'div' elements found
```

```python
# Check the total number of 'div' elements
len(div)
```

```python
# Find all 'div' elements with class "title" (assuming this is the correct class for titles)
mydivs = soup.find_all("div", {"class": "title"})
mydivs  # Displays all 'div' elements with class "title"
```

```python
# Find all 'div' elements with class "price" (assuming this is the correct class for prices)
pricedivs = soup.find_all("div", {"class": "price"})
pricedivs  # Displays all 'div' elements with class "price"
```

```python
# Find all 'div' elements with class "location" (assuming this is the correct class for locations)
locdivs = soup.find_all("div", {"class": "location"})
locdivs  # Displays all 'div' elements with class "location"
```

```python
# Check the legth for each
print(len(mydivs))
print(len(pricedivs))
print(len(locdivs))
```

```python
# Initialize an empty list to store extracted data
data = []

# Loop through the elements based on the number of titles found (assuming all lists are the same length)
for i in range(len(mydivs)):
    # Extract text from each div and remove any extra whitespace
    title = mydivs[i].text.strip() if mydivs[i] else 'N/A'  # Get the title text or 'N/A' if not available
    price = pricedivs[i].text.strip() if pricedivs[i] else 'N/A'  # Get the price text or 'N/A' if not available
    location = locdivs[i].text.strip() if locdivs[i] else 'N/A'  # Get the location text or 'N/A' if not available

    # Append the extracted data to the list as a dictionary
    data.append({'Title': title, 'Price': price, 'Location': location})
```

```python
# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the DataFrame to verify the extracted data
print(df.head())
```
